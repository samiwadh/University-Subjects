## Most common `read_csv` options

*  `sep` or `delimiter`: the string that separates the fields. 
*  `delim_whitespace`: a boolean that states if any whitespace is the delimiter
*  `skiprows`: lines to skip. Usually at the beginning of the file, if it contains a comment
*  `nrows`: number of rows to read. Useful for debugging on huge files
*  `low_memory`: boolean (default `True`). Sometimes messes up parsing
*  `na_values`, `keep_default_na`, `na_filter`: change the default behavior for parsing missing data
*  `parse_dates`: dates are a mess to parse. State which columns contain dates
*  `infer_datetime_format`, `dayfirst`: may speed up parsing dates
*  `thousands`, `decimal`: hints for parsing non US-based numbers
*  `encoding`: sometimes you have to set it to `utf-8`

```python


import pandas as pd 
orders = pd.read_csv('Dataset/Orders.csv', low_memory=False)
# Helps with big files; setting low_memory=False can fix mixed-type column errors
orders.head()
```

```python
import pandas as pd
orders = pd.read_csv('Dataset/Orders.csv', na_values=['?', 'N/A', 'Missing'])
# Treat these strings as missing (NaN) values
orders.head()
```

```python
import pandas as pd
orders = pd.read_csv('Dataset/Orders.csv', keep_default_na=False)
# Disable pandas' default NA strings (like "NaN", "NA")
orders.head()
```

```python
import pandas as pd
orders = pd.read_csv('Dataset/Orders.csv', na_filter=False)
# Turn off missing-value detection (faster for clean data)
orders.head()
```

```python
import pandas as pd
orders = pd.read_csv('Dataset/Orders.csv', parse_dates=['OrderDate'])
# Automatically convert the 'Order Date' column to datetime
orders.head()
```

```python
import pandas as pd
orders = pd.read_csv('Dataset/Orders.csv', parse_dates=['OrderDate'], infer_datetime_format=True)
# Speed up date parsing when pandas can guess the format
orders.head()
```

```python
import pandas as pd
orders = pd.read_csv('Dataset/Orders.csv', parse_dates=['OrderDate'], dayfirst=True)
# Interpret dates like 31/12/2024 as Dec 31 (European style)
orders.head()
```

```python
import pandas as pd
orders = pd.read_csv('Dataset/Orders.csv', thousands=',')
# Handle numbers like 1,000 or 50,000 properly
orders.head()
```

```python
import pandas as pd
orders = pd.read_csv('Dataset/Orders.csv', decimal=',')
# Handle European decimals (e.g. 3,14 instead of 3.14)
orders.head()
```

```python
import pandas as pd
orders = pd.read_csv('Dataset/Orders.csv', encoding='utf-8')
# Fix encoding errors (try 'latin1' or 'ISO-8859-1' if UTF-8 fails)
orders.head()
```
